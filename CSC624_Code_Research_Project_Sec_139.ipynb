```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

def plot_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()


# لعمل التحليلات الاحصائية لعمل المودل
```

### The First Modle to Detected SMILE is Active or InActive

```python
arguments = [
    '--data_path', '/content/drive/MyDrive/yousefProject/class_prep.csv',
    '--dataset_type', 'classification',
    '--save_dir', 'test_checkpoints_cl',
    '--epochs', '10',
    '--save_smiles_splits'
]

args = chemprop.args.TrainArgs().parse_args(arguments)
mean_score, std_score = chemprop.train.cross_validate(args=args, train_func=chemprop.train.run_training)


# تفعيل المودل و جعله يتدرب على البيانات المجهزه مسبقا
```

```python
arguments = [
    '--test_path', 'test_checkpoints_cl/fold_0/test_smiles.csv',
    '--preds_path', 'test_preds_cl.csv',
    '--checkpoint_dir', 'test_checkpoints_cl'
]

args = chemprop.args.PredictArgs().parse_args(arguments)
preds = chemprop.train.make_predictions(args=args)

# اختبار المودل على البيانات المقسمه مسبقا من قبل المودل حيث يتم جعل ثمانين بالمية من البيانات تدريب و عشره بالمية تحقق و عشرة بالمية المتبقية اختبار
```

```python
df = pd.read_csv('test_checkpoints_cl/fold_0/test_full.csv')
df['preds'] = [x[0] for x in preds]
df

# قرائة البينات التي تم التنبؤ بها من قبل المودل
```

```python
# Convert predictions to binary (0 or 1) if they are probabilities
df['preds'] = df['preds'].apply(lambda x: 1 if x > 0.5 else 0)

# Plot confusion matrix
plot_confusion_matrix(df['state'], df['preds'])


# وضع شرط معين اذا كان اقل من 5 اعتبره صفر واذا كان اكبر من 5 خليه 1
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(df['state'], df['preds'])

total_samples = cm.sum()
correct_predictions = cm.trace()  # Sum of diagonal elements
incorrect_predictions = total_samples - correct_predictions

# Assuming binary classification (classes 0 and 1)
tp = cm[0][0]
tn = cm[1][1]
fp = cm[0][1]
fn = cm[1][0]

# Print results (modify for your specific needs)
print(f"Total Samples: {total_samples}")
print(f"Correct Predictions: {correct_predictions}")
print(f"Incorrect Predictions: {incorrect_predictions}")
print(f"True Positives (TP): {tp}")
print(f"True Negatives (TN): {tn}")
print(f"False Positives (FP): {fp}")
print(f"False Negatives (FN): {fn}")


# Calculate Accuracy
accuracy = (tp + tn) / (tp + tn + fp + fn)

# Calculate Precision
precision = tp / (tp + fp)

# Calculate Recall
recall = tp / (tp + fn)

# Calculate F1-score
f1 = 2 * (precision * recall) / (precision + recall)

# Print the results
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")


# كود يقيس مدى فاعلية المودل بناء على البيانات التي تم الاختبار عليها
```
